```python
# We are reading the csv and printing it out
df = pd.read_csv('data.csv', parse_dates=['START DATE'], index_col='START DATE')
df.head()
```

```python
# Shows the data types
df.info()
```

```python
# Plot 1: Usage plot over the months
def plot_plots(df, x, y, title="", xlabel='', ylabel='', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='red', label='Electricity')
    #plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    #plt.plot(x, z, color='red', label='Divorce')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    #plt.xticks(rotation=90)
    
    #plt.show()
# Here, the two datdaframes are combined
#df_two = pd.concat([df, df_smallcols])
plot_plots(df, x=df.index, y=df.USAGE, title='Amt of Electricity')
```
